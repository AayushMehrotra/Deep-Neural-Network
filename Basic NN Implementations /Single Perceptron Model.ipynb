## Getting Best Fit equation

#### Ax + bY + c = 0
#### Y = -(Ax + c)/b

# Matrix Solution

# Prediction Matrix

```python
def vector_predict(x, w):
    dot = x.dot(w)

    if dot > 0:
        return 1
    else:
        return 0
```

# Get Weight Matrix

```python
def get_weight_vector(x , w , y ,iterate):
    rate = 0.1
    weights = []
    for j in range(iterate):
        square_error = 0
        for i in range(len(x)):
            predict = vector_predict(x[i], w)
            error = y[i] - predict
            square_error += error**2
            w = w + rate * error * x[i]

            
        weights.append(w)
        print('Epoch number ',j,'Error ',square_error)
        if square_error == 0:
            break
    return weights
```

# Get Line

```python
def get_line(wt, x1, x2):
    y = []
    ys = []
    for weight in wt:
        for w in range(len(weight)-1):
            w1 = weight[1]
            w2 = weight[2]
            bias = weight[0]
            y1 = -(w1 * x1 + bias)/w2
            y2 = -(w1 * x2 + bias)/w2
        y.append(y1)
        y.append(y2)
        ys.append(y)
        y = []
    return ys
```

# DATASET 1

## W = W + rate * error * X

## error = expected - predicted

## error = Yactual - Ypredict

## Creating X matrix (1, X1, X2)

```python
x = np.ones(np.shape(dataset1))
x[:, 1] = dataset1.X1
x[:, 2] = dataset1.X2
```

# Creating W matix (Bias, W1, W2)

```python
w = np.random.random_sample(3)
```

# Creating Yactual(Expected Matrix)

```python
y = dataset1.Y
y = np.array(y)
```

# Getting the linear line

```python
weights = get_weight_vector(x, w, y, 20)
x_min = min(dataset1['X1'])
x_max = max(dataset1['X1'])
y_points = y_points = get_line(weights,x_min,x_max)
y1 = y_points[len(y_points) - 1][0]
y2 = y_points[len(y_points) - 1][1]
```

# Normal Plot

```python
%matplotlib inline
plt.scatter(x = dataset1['X1'], y = dataset1['X2'], c = dataset1['Y'])
plt.plot([x_min, x_max], [y1, y2], color='k', linestyle='-', linewidth=2)
plt.show()
```

# Animated Plot

```python
%matplotlib nbagg
fig, ax = plt.subplots()
length = len(y_points)
x = [x_min, x_max]
line, = ax.plot(x, y_points[0])
def animate(i):
    line.set_ydata(y_points[i])   # update the data
    return line,


# Init only required for blitting to give a clean slate.
def init():
    line.set_ydata(np.ma.array(x, mask=True))
    return line,
ani = animation.FuncAnimation(fig, animate, np.arange(1, length), init_func=init,
    interval=25, blit=True)
plt.scatter(x = dataset1['X1'], y = dataset1['X2'], c = dataset1['Y'])
plt.show()
```
